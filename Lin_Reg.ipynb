# Linear Regression

## import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
from sklearn.model_selection import train_test_split
%matplotlib inline
```

### importind datasets and extracting independent and dependent variable


```python
companies=pd.read_csv("C:\\Users\\dhruv\\Downloads\\1000_Companies.csv")
x=companies.iloc[:, :-1].values #independent [features]
y=companies.iloc[:,4].values #dependent [labels]
```

```python
companies.head(4)
```

## to convert categorical data[city] into num data

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
label_encoder=LabelEncoder()
x[:,3]=label_encoder.fit_transform(x[:,3])

```

```python
one_hot_encoder=OneHotEncoder()
```

### avoid dummmy variable loop

```python
x=x[:,1:]
```

## splitting data

```python
X_train, X_test, y_train, y_test=train_test_split(x,y,test_size=0.2, random_state=0)
```

## fitting multiple linear regression 

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train, y_train)
```

### predict the test 

```python
y_pred=reg.predict(X_test)
y_pred
```

```python
reg.coef_
```

```python
reg.intercept_
```

### calculating the score 

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)*100
```
